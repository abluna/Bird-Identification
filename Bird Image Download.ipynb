```python
from bing_image_downloader import downloader
import os
import pandas as pd
```

```python
## Bring in bird list
bird_df = pd.read_csv('C:/Users/abbal/Documents/Data Science Personal Projects/Py - Bird Classification/list_of_san_diego_birds.csv')
```

```python
## Set up list of birds
bird_list = bird_df.iloc[0:25:,0]

for bird in bird_list:
    
    downloader.download(bird + " Bird", 
                        limit=50,
                        output_dir='C:/Users/abbal/Documents/Data Science Personal Projects/Py - Bird Classification/Raw Images', 
                        adult_filter_off=True, 
                        force_replace=False, 
                        timeout=60,
                        resize_dim=(256, 256),
                        verbose=False)
```
